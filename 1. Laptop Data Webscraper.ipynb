```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

#laptop links webscraper
#first 20 pages
laptoplinks = []
for x in range(1,20):
    r = requests.get(f'https://pricebaba.com/laptop/pricelist/all-laptops-sold-in-india?page={x}')
    soup = BeautifulSoup(r.content,'html.parser')
    laptoplist = soup.find_all('div',class_='col-3 p-h-xs v-al-top flt-l')
   
    for item in laptoplist:
        for link in item.find_all('a',href=True):
            laptoplinks.append(link['href'])
print(len(laptoplinks))
```
